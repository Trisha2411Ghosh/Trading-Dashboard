```python
def generate_unique_trade_ids(n, length=8):
    trade_ids = set()  # To store unique trade IDs

    # Character set: uppercase letters and digits
    charset = string.ascii_uppercase + string.digits

    # Continue generating until we have the required number of unique IDs
    while len(trade_ids) < n:
        trade_id = ''.join(random.choices(charset, k=length))
        trade_ids.add(trade_id)

    return list(trade_ids)

# Generate 40 million unique trade IDs
trade_ids = generate_unique_trade_ids(3000000)
```

```python
existing_instruments = trade_df['instrument_id'].unique()
```

```python
new_data = {
    "trade_id": trade_ids,
    "instrument_id": [random.choice(existing_instruments) for _ in range(3000000)]
}

# Step 3: Create the new DataFrame with trade IDs and instrument IDs
new_trade_info_df = pd.DataFrame(new_data)
```

```python
new_trade_info_df_sorted = new_trade_info_df.sort_values(by='instrument_id')
```

```python
new_trade_info_df_sorted['traded_volume_lakhs'] = [round(random.uniform(10, 500), 2) for _ in range(len(new_trade_info_df))]
```

```python
new_trade_info_df_sorted['traded_value_cr'] = new_trade_info_df_sorted['traded_volume_lakhs'] * [round(random.uniform(0.5, 1.5), 2) for _ in range(len(new_trade_info_df))]
```

```python
new_trade_info_df_sorted['impact_cost'] = [round(random.uniform(0.1, 0.5), 2) for _ in range(len(new_trade_info_df))]
```

```python
instrument_ffm_cap_map = trade_df.set_index('instrument_id')['ffm_cap'].to_dict()
```

```python
new_trade_info_df_sorted['ffm_cap'] = new_trade_info_df_sorted['instrument_id'].map(instrument_ffm_cap_map)
```

```python
new_trade_info_df_sorted['total_market_cap_cr'] = new_trade_info_df_sorted['ffm_cap'].apply(
    lambda x: round(x * (1 + ((random.random() - 0.5) * 0.1)), 2)
)
```

```python
new_trade_info_df_sorted['percent_deliverable_traded_quantity'] = [round(random.uniform(20, 80), 2) for _ in range(len(new_trade_info_df))]
```

```python
new_trade_info_df_sorted['applicable_margin_rate'] = [round(random.uniform(5, 15), 2) for _ in range(len(new_trade_info_df))]
```

```python
new_trade_info_df_sorted['face_value'] = [random.choice([1, 10]) for _ in range(len(new_trade_info_df))]
```

```python
duplicates = new_trade_info_df_sorted[new_trade_info_df_sorted.duplicated(subset='trade_id', keep=False)]

if not duplicates.empty:
    print("Duplicate trade_ids found:")
    print(duplicates)
else:
    print("No duplicate trade_ids found.")
```
