# Trade ID Generation

This code snippet demonstrates how to generate a list of unique trade IDs. It uses a prefix ("T") and a sequence number, formatted with leading zeros, to create 3 million trade IDs.

---

### Code Breakdown:

1. **`prefix = "T"`**:
   - The prefix for each trade ID. In this case, each ID will start with the letter "T".
   
2. **`count = 3_000_000`**:
   - The total number of trade IDs to generate. Here, the count is set to 3 million.

3. **Trade ID Generation**:
   - The list comprehension iterates over a range of numbers from 1 to `count` (inclusive). For each iteration, it generates a string formatted as `"T"` followed by a 7-digit number (padded with leading zeros, if necessary).

   - **`f"{prefix}{i:07}"`**:
     - `prefix`: Adds the "T" at the beginning of each ID.
     - `{i:07}`: This uses Python's f-string formatting to ensure that each number `i` is padded with leading zeros to always be 7 digits long (e.g., `0000001`, `0000002`, ..., `3000000`).

4. **Output**:
   - The result is a list of 3 million trade IDs, starting from `T0000001` and endingfor i in range(1, count + 1)]


```python
prefix = "T"
count = 3_000_000  # 3 million

# Generate trade IDs
trade_ids = [f"{prefix}{i:07}" for i in range(1, count + 1)]
```

# Random Date Generation and Assignment

This code snippet demonstrates how to generate random dates within a specified range and assign them to a new column in a DataFrame.

---

### Code Breakdown:

1. **`random_date(start, end)`**:
   - **`start`**: The starting date of the range.
   - **`end`**: The ending date of the range.
   - The function calculates the difference (`delta`) between the `end` and `start` dates.
   - It then generates a random number of days (`random_days`) between 0 and the total number of days in the `delta`.
   - Finally, it returns a new date that is the `start` date plus the `random_days`, generating a random date within the specified range.

2. **`new_trade_info_df_sorted['trade_date']`**:
   - A new column named `'trade_date'` is added to the `new_trade_info_df_sorted` DataFrame.
   - The list comprehension `[random_date(start_date, end_date) for _ in range(len(new_trade_info_df_sorted))]` is used to generate a random date for each row in the DataFrame.
   - The `random_date` function is called for each row, with the `start_date` and `end_date` parameters defining the range for the random date.

---




```python
def random_date(start, end):
    delta = end - start
    random_days = np.random.randint(0, delta.days + 1)
    return start + timedelta(days=random_days)

# Add a 'date' column with random dates
new_trade_info_df_sorted['trade_date'] = [random_date(start_date, end_date) for _ in range(len(new_trade_info_df_sorted))]
```

# Connecting to PostgreSQL Database with psycopg2

This code demonstrates how to establish a connection to a PostgreSQL database using the `psycopg2` library and execute a simple query to verify the connection.

---

### Code Breakdown:

1. **Connection URL**:
   - The connection URL contains the following parameters:
     - `trade_owner`: Username
     - `Password`: Password
     - `ep-tiny-fire-a1325guv.ap-southeast-1.aws.neon.tech`: Host
     - `trade`: Database name
     - `sslmode=require`: Enforces SSL encryption for the connection.

2. **Database Connection**:
   - `psycopg2.connect(connection_url)` is used to create a connection to the PostgreSQL database using the connection string.
   - If the connection is successful, a message "Connection successful!" is printed.

3. **Creating a Cursor**:
   - The cursor is created using `conn.cursor()`. This allows you to interact with the database and execute SQL queries.

4. **Executing SQL Query**:
   - The query `cursor.execute("SELECT version();")` fetches the version of the PostgreSQL database to verify that the connection is active and working.
   - The result is fetched using `cursor.fetchone()` and printed.

5. **Closing Connection**:
   - Both the cursor and the connection are closed with `cursor.close()` and `conn.close()` after the query execution.

---


```python
import psycopg2

# Database connection parameters
connection_url = "postgresql://trade_owner:password@ep-tiny-fire-a1325guv.ap-southeast-1.aws.neon.tech/trade?sslmode=require"

try:
    # Connect to the PostgreSQL database
    conn = psycopg2.connect(connection_url)
    print("Connection successful!")

    # Create a cursor to execute SQL queries
    cursor = conn.cursor()

    # Example query to check connection
    cursor.execute("SELECT version();")
    db_version = cursor.fetchone()
    print("Database version:", db_version)

    # Close the cursor and connection
    cursor.close()
    conn.close()

except Exception as e:
    print("Error connecting to the database:", e)
```

# Inserting Data into Database in Batches

This code demonstrates how to insert a large dataset into a PostgreSQL database in smaller batches. The first batch (comprising the first 1 million rows) is selected from a DataFrame and inserted into the database.

---

### Code Breakdown:

1. **Selecting the First Batch**:
   - The DataFrame `new_trade_info_df_sorted` is sliced to select the first 1 million rows using `iloc[:1_000_000]`. This creates a subset of the DataFrame (`batch_df`) that includes the rows indexed from 0 to 999,999.
   
2. **Inserting the Data into the Database**:
   - The `to_sql()` method is used to insert the `batch_df` DataFrame into the specified database table (`table_name`).
     - `engine` is the SQLAlchemy engine used for database connection.
     - `if_exists='append'`: If the table already exists, the data will be appended to it.
     - `index=False`: The index of the DataFrame will not be inserted as a separate column.

3. **Error Handling**:
   - The operation is wrapped in a `try-except` block to catch and display any errors that may occur during the batch insertion.
   - If the insertion is successful, a message "First batch (0 to 1,000,000) inserted successfully" is printed.
   - If an error occurs, the exception is caught, and an error message is displayed with details.
4. **Repetition**:
   - Run the same code again by modifying the line `batch_df = new_trade_info_df_sorted.iloc[:1_000_000]`  to `batch_df =        new_trade_info_df_sorted.iloc[1_000_000:2_000_000]` and so on 
---



```python
try:
    # Select the first 1 million rows (0 to 999,999)
    batch_df = new_trade_info_df_sorted.iloc[:1_000_000]

    # Insert the batch into the database
    batch_df.to_sql(table_name, engine, if_exists='append', index=False)

    print("First batch (0 to 1,000,000) inserted successfully.")

except Exception as e:
    print("Error inserting first batch (0 to 1,000,000):", e)
```
